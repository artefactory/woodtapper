```python
from collections import Counter
import os
import sys
#sys.path.insert(1, os.path.abspath(os.path.join(os.getcwd(), os.pardir)))
os.chdir('../')
import time

import numpy as np

import skgrf
from skgrf.ensemble import GRFForestClassifier
```

## Load data set

Make sure you have already generated the dat set trough the first chunks from timing-exampleSampling-with-skgrf-plot.py

```python
output_dir_data_set = "sim-data-time" #simulation data for timing

X_train_original = np.genfromtxt(os.path.join(output_dir_data_set, "X_train.csv"), delimiter = ",")
y_train_original = np.genfromtxt(os.path.join(output_dir_data_set, "y_train.csv"), delimiter = ",")
X_test_original = np.genfromtxt(os.path.join(output_dir_data_set, "X_test.csv"), delimiter = ",")
y_test_original = np.genfromtxt(os.path.join(output_dir_data_set, "y_test.csv"), delimiter = ",")
```

## Timing by samples:

```python
output_dir_ExSampling = "exp-ExSampling/" #simulation data for example sampling
```

```python
list_n_samples_test = [1000, 2500, 5000, 7500, 10000, 25000, 50000, 75000, 100000]
```

```python
n_samples_train = 1000
n_samples_test = 50000
n_dim = 20

X_train = X_train_original[:n_samples_train,:n_dim]
y_train = y_train_original[:n_samples_train]

X_test = X_test_original[:n_samples_test,:n_dim] 
y_test = y_test_original[:n_samples_test]
```

```python
grf_clf = GRFForestClassifier(n_estimators=100,mtry=10,
                              sample_fraction=0.99,ci_group_size=1,alpha=0.001,
                              honesty=True,honesty_fraction=0.5,honesty_prune_leaves=False,
                              seed=0,n_jobs=1)
grf_clf.fit(X_train, y_train)
```

```python
##### SAMPLES ####
list_time = []  
for n_samples_test in list_n_samples_test:
    X_test_subset = X_test_original[:n_samples_test,:n_dim]
    time_start = time.time()
    weights = grf_clf.get_kernel_weights(X_test_subset,oob_prediction=False)
    time_end = time.time()
    time_ = time_end - time_start
    list_time.append(time_)
    print("Get kernel time for n_samples_test =", n_samples_test, ": ", time_)

np.save(os.path.join(output_dir_ExSampling, "list_time_samples_skgrf_1.npy"), np.array(list_time))
list_time  
```

```python
##### SAMPLES 2 ####
list_time_2 = []  
for n_samples_test in list_n_samples_test:
    X_test_subset = X_test_original[:n_samples_test,:n_dim]
    time_start = time.time()
    weights = grf_clf.get_kernel_weights(X_test_subset,oob_prediction=False)
    time_end = time.time()
    time_ = time_end - time_start
    list_time_2.append(time_)
    print("Get kernel time for n_samples_test =", n_samples_test, ": ", time_)
np.save(os.path.join(output_dir_ExSampling, "list_time_samples_skgrf_2.npy"), np.array(list_time_2))
  
```

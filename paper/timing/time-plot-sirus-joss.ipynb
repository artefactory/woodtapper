## Samples

```python
time_python_samples = np.load(os.path.join(output_dir_Rules, "list_time_samples_python.npy"), allow_pickle=True).T
time_r_samples = np.loadtxt(os.path.join(output_dir_Rules, "list_time_samples_r.csv"), delimiter=',', skiprows=1).T.reshape(5,5)
time_julia_samples = np.loadtxt(os.path.join(output_dir_Rules, "list_time_samples_julia.csv"), delimiter=',')
```

```python
GrowForest_samples_python = np.loadtxt(os.path.join(output_dir_Rules, "list_GrowForest_samples_python.csv"), delimiter=',')
GrowForest_samples_r = np.loadtxt(os.path.join(output_dir_Rules, "list_GrowForest_samples_r.csv"), delimiter=',')
GrowForest_samples_julia = np.loadtxt(os.path.join(output_dir_Rules, "list_GrowForest_samples_julia.csv"), delimiter=',')
```

```python
RulesExtraction_samples_python = np.loadtxt(os.path.join(output_dir_Rules, "list_RulesExtraction_samples_python.csv"), delimiter=',')
RulesExtraction_samples_r = np.loadtxt(os.path.join(output_dir_Rules, "list_RulesExtraction_samples_r.csv"), delimiter=',')
RulesExtraction_samples_julia = np.loadtxt(os.path.join(output_dir_Rules, "list_RulesExtraction_samples_julia.csv"), delimiter=',')
```

```python

## Pyhton

data = {
    "n_samples_train": [100000, 200000, 300000, 400000, 500000],
    
    # Fit time
    "Fit_time_Run1": time_python_samples[0],
    "Fit_time_Run2": time_python_samples[1],
    "Fit_time_Run3": time_python_samples[2],
    "Fit_time_Run4": time_python_samples[3],
    "Fit_time_Run5": time_python_samples[4],

    # Grow forest
    "Grow_forest_Run1": GrowForest_samples_python[0],
    "Grow_forest_Run2": GrowForest_samples_python[1],
    "Grow_forest_Run3": GrowForest_samples_python[2],
    "Grow_forest_Run4": GrowForest_samples_python[3],
    "Grow_forest_Run5": GrowForest_samples_python[4],

    # Rules extraction
    "Rules_extraction_Run1": RulesExtraction_samples_python[0],
    "Rules_extraction_Run2": RulesExtraction_samples_python[1],
    "Rules_extraction_Run3": RulesExtraction_samples_python[2],
    "Rules_extraction_Run4": RulesExtraction_samples_python[3],
    "Rules_extraction_Run5": RulesExtraction_samples_python[4],
}

df_py = pd.DataFrame(data)

# Optional: compute mean and std per n_samples_train
df_py["Fit_time_mean"] = df_py[[f"Fit_time_Run{i}" for i in range(1,6)]].mean(axis=1)
df_py["Grow_forest_mean"] = df_py[[f"Grow_forest_Run{i}" for i in range(1,6)]].mean(axis=1)
df_py["Rules_extraction_mean"] = df_py[[f"Rules_extraction_Run{i}" for i in range(1,6)]].mean(axis=1)

df_py["Fit_time_std"] = df_py[[f"Fit_time_Run{i}" for i in range(1,6)]].std(axis=1)
df_py["Grow_forest_std"] = df_py[[f"Grow_forest_Run{i}" for i in range(1,6)]].std(axis=1)
df_py["Rules_extraction_std"] = df_py[[f"Rules_extraction_Run{i}" for i in range(1,6)]].std(axis=1)

df_py
```

```python
## R
data = {
    "n_samples_train": [100000, 200000, 300000, 400000, 500000],

    # Fit time
    "Fit_time_Run0": time_r_samples[0],
    "Fit_time_Run1": time_r_samples[1],
    "Fit_time_Run2": time_r_samples[2],
    "Fit_time_Run3": time_r_samples[3],
    "Fit_time_Run4": time_r_samples[4],

    # Grow forest
    "Grow_forest_Run0": GrowForest_samples_r[0],
    "Grow_forest_Run1": GrowForest_samples_r[1],
    "Grow_forest_Run2": GrowForest_samples_r[2],
    "Grow_forest_Run3": GrowForest_samples_r[3],
    "Grow_forest_Run4": GrowForest_samples_r[4],

    # Rules extraction
    "Rules_extraction_Run0": RulesExtraction_samples_r[0],
    "Rules_extraction_Run1": RulesExtraction_samples_r[1],
    "Rules_extraction_Run2": RulesExtraction_samples_r[2],
    "Rules_extraction_Run3": RulesExtraction_samples_r[3],
    "Rules_extraction_Run4": RulesExtraction_samples_r[4],
}

df_r = pd.DataFrame(data)

# Optional: compute mean and std across runs
df_r["Fit_time_mean"] = df_r[[f"Fit_time_Run{i}" for i in range(5)]].mean(axis=1)
df_r["Grow_forest_mean"] = df_r[[f"Grow_forest_Run{i}" for i in range(5)]].mean(axis=1)
df_r["Rules_extraction_mean"] = df_r[[f"Rules_extraction_Run{i}" for i in range(5)]].mean(axis=1)

df_r["Fit_time_std"] = df_r[[f"Fit_time_Run{i}" for i in range(5)]].std(axis=1)
df_r["Grow_forest_std"] = df_r[[f"Grow_forest_Run{i}" for i in range(5)]].std(axis=1)
df_r["Rules_extraction_std"] = df_r[[f"Rules_extraction_Run{i}" for i in range(5)]].std(axis=1)

df_r
```

```python
#Julia
data = {
    "run": [0]*5 + [1]*5 + [2]*5 + [3]*5 + [4]*5,
    "n_samples": [100000, 200000, 300000, 400000, 500000]*5,
    "training_time": time_julia_samples.ravel().tolist(),
    "grow_forest_time": GrowForest_samples_julia.ravel().tolist(),
    "rules_extraction_time": RulesExtraction_samples_julia.ravel().tolist()
}

df_julia = pd.DataFrame(data)

# Optional: compute mean times across runs for each n_samples
df_julia_summary_mean = df_julia.groupby("n_samples")[["training_time", "grow_forest_time", "rules_extraction_time"]].mean().reset_index()
df_julia_summary_std = df_julia.groupby("n_samples")[["training_time", "grow_forest_time", "rules_extraction_time"]].std().reset_index()

df_julia_summary_mean
```

```python
# Sample 
n_samples = [100000, 200000, 300000, 400000, 500000]

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
#fig.suptitle(r"Computation Time for $n$=300K and $M$=1000", fontsize=20)


# --- Subplot 1: Fit Time ---
axes[0].plot(n_samples, df_py["Fit_time_mean"].values, 'o--', label='Python (Ours)',c='royalblue')
axes[0].fill_between(n_samples, df_py["Fit_time_mean"].values-df_py["Fit_time_std"].values,
                     df_py["Fit_time_mean"].values+df_py["Fit_time_std"].values,alpha=0.5,color='skyblue')
axes[0].plot(n_samples, df_r["Fit_time_mean"].values, 's--', label='R',c='crimson')
axes[0].fill_between(n_samples, df_r["Fit_time_mean"].values-df_r["Fit_time_std"].values,
                     df_r["Fit_time_mean"].values+df_r["Fit_time_std"].values,alpha=0.5,color='lightcoral')
axes[0].plot(n_samples, df_julia_summary_mean["training_time"].values, '^--', label='Julia',c="green")
axes[0].fill_between(n_samples, df_julia_summary_mean["training_time"].values-df_julia_summary_std["training_time"].values,
                     df_julia_summary_mean["training_time"].values+df_julia_summary_std["training_time"].values,alpha=0.2,color='limegreen')
axes[0].set_yscale('log')

# --- Y-axis log scale with powers of 10 ---
axes[0].set_yscale('log')
axes[0].yaxis.set_major_locator(LogLocator(base=10.0))       # ticks at 10^n
axes[0].yaxis.set_major_formatter(ScalarFormatter())          # show as 10^n
axes[0].yaxis.get_offset_text().set_visible(False)            # remove offset text
# --- X-axis linear but show powers of 10 ---
x_ticks = [1e5, 2e5, 3e5, 4e5, 5e5]                      # linear spacing
axes[0].set_xticks(x_ticks)
axes[0].set_xticklabels([f'${i+1}$'+r"$\times$"+f'$10^{int(np.log10(tick))}$' for i,tick in enumerate(x_ticks)])  # show powers

axes[0].set_title("Total",fontsize=18)
axes[0].set_xlabel("Number of samples",fontsize=15)
axes[0].set_ylabel("Time (s)",fontsize=15)
axes[0].legend(fontsize=12)
axes[0].grid(False)

# --- Subplot 2: Rules Extraction ---
axes[1].plot(n_samples, df_py["Rules_extraction_mean"].values, 'o--', label='Python (Ours)',c='royalblue')
axes[1].fill_between(n_samples, df_py["Rules_extraction_mean"].values-df_py["Rules_extraction_std"].values,
                     df_py["Rules_extraction_mean"].values+df_py["Rules_extraction_std"].values,alpha=0.5,color='skyblue')
axes[1].plot(n_samples, df_r["Rules_extraction_mean"].values, 's--', label='R',c='crimson')
axes[1].fill_between(n_samples, df_r["Rules_extraction_mean"].values-df_r["Rules_extraction_std"].values,
                     df_r["Rules_extraction_mean"].values+df_r["Rules_extraction_std"].values,alpha=0.5,color='lightcoral')
axes[1].plot(n_samples, df_julia_summary_mean["rules_extraction_time"].values, '^--', label='Julia',c="green")
axes[1].fill_between(n_samples, df_julia_summary_mean["rules_extraction_time"].values-df_julia_summary_std["rules_extraction_time"].values,
                     df_julia_summary_mean["rules_extraction_time"].values+df_julia_summary_std["rules_extraction_time"].values,alpha=0.2,color='limegreen')

# --- Y-axis log scale with powers of 10 ---
axes[1].set_yscale('log')
axes[1].yaxis.set_major_locator(LogLocator(base=10.0))       # ticks at 10^n
axes[1].yaxis.set_major_formatter(ScalarFormatter())          # show as 10^n
axes[1].yaxis.get_offset_text().set_visible(False)            # remove offset text
# --- X-axis linear but show powers of 10 ---
x_ticks = [1e5, 2e5, 3e5, 4e5, 5e5]                      # linear spacing
axes[1].set_xticks(x_ticks)
axes[1].set_xticklabels([f'${i+1}$'+r"$\times$"+f'$10^{int(np.log10(tick))}$' for i,tick in enumerate(x_ticks)])  # show powers
axes[1].set_title("Rules Extraction",fontsize=18)
axes[1].set_xlabel("Number of samples",fontsize=15)
axes[1].set_ylabel("Time (s)",fontsize=15)
axes[1].legend(fontsize=12)
axes[1].grid(False)

# --- Subplot 3: Grow Forest Time ---
axes[2].plot(n_samples, df_py["Grow_forest_mean"].values, 'o--', label='Python (Ours)',c='royalblue')
axes[2].fill_between(n_samples, df_py["Grow_forest_mean"].values-df_py["Grow_forest_std"].values,
                     df_py["Grow_forest_mean"].values+df_py["Grow_forest_std"].values,alpha=0.5,color='skyblue')
axes[2].plot(n_samples, df_r["Grow_forest_mean"].values, 's--', label='R',c='crimson')
axes[2].fill_between(n_samples, df_r["Grow_forest_mean"].values-df_r["Grow_forest_std"].values,
                     df_r["Grow_forest_mean"].values+df_r["Grow_forest_std"].values,alpha=0.5,color='lightcoral')
axes[2].plot(n_samples, df_julia_summary_mean["grow_forest_time"].values, '^--', label='Julia',c="green")
axes[2].fill_between(n_samples, df_julia_summary_mean["grow_forest_time"].values-df_julia_summary_std["grow_forest_time"].values,
                     df_julia_summary_mean["grow_forest_time"].values+df_julia_summary_std["grow_forest_time"].values,alpha=0.2,color='limegreen')
# --- Y-axis log scale with powers of 10 ---
axes[2].set_yscale('log')
axes[2].yaxis.set_major_locator(LogLocator(base=10.0))       # ticks at 10^n
axes[2].yaxis.set_major_formatter(ScalarFormatter())          # show as 10^n
axes[2].yaxis.get_offset_text().set_visible(False)            # remove offset text
# --- X-axis linear but show powers of 10 ---
x_ticks = [1e5, 2e5, 3e5, 4e5, 5e5]                      # linear spacing
axes[2].set_xticks(x_ticks)
axes[2].set_xticklabels([f'${i+1}$'+r"$\times$"+f'$10^{int(np.log10(tick))}$' for i,tick in enumerate(x_ticks)])  # show powers
axes[2].set_title("Fit Forest",fontsize=18)
axes[2].set_xlabel("Number of samples",fontsize=15)
axes[2].set_ylabel("Time (s)",fontsize=15)
axes[2].legend(fontsize=12)
axes[2].grid(False)


plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig(os.path.join(output_dir_Rules,"run-time-samples-log-5threads-final.pdf"))
plt.show()
```

```python
GrowForest_dims_python = np.loadtxt(os.path.join(output_dir_Rules, "list_GrowForest_dims_python.csv"), delimiter=',')
GrowForest_dims_r = np.loadtxt(os.path.join(output_dir_Rules, "list_GrowForest_dims_r.csv"), delimiter=',')
GrowForest_dims_julia = np.loadtxt(os.path.join(output_dir_Rules, "list_GrowForest_dims_julia.csv"), delimiter=',')
```

```python
RulesExtraction_dims_python = np.loadtxt(os.path.join(output_dir_Rules, "list_RulesExtraction_dims_python.csv"), delimiter=',')
RulesExtraction_dims_r = np.loadtxt(os.path.join(output_dir_Rules, "list_RulesExtraction_dims_r.csv"), delimiter=',')
RulesExtraction_dims_julia = np.loadtxt(os.path.join(output_dir_Rules, "list_RulesExtraction_dims_julia.csv"), delimiter=',')
```

```python
#Python
data = {
    "run": [1]*9 + [2]*9 + [3]*9 + [4]*9 + [5]*9,
    "n_dim": [15, 25, 50, 75, 100, 125, 150, 175, 200]*5,
    
    # Fit times
    "fit_time": time_python_dims.ravel().tolist(),
    
    # Grow forest times
    "grow_forest_time": GrowForest_dims_python.ravel().tolist(),
    
    # Rules extraction times
    "rules_extraction_time": RulesExtraction_dims_python.ravel().tolist(),
}

# Create dataframe
df_python = pd.DataFrame(data)

# Compute mean times across runs for each n_dim
df_py_summary_mean = (
    df_python
    .groupby("n_dim")[["fit_time", "grow_forest_time", "rules_extraction_time"]]
    .mean()
    .reset_index()
)
df_py_summary_std = (
    df_python
    .groupby("n_dim")[["fit_time", "grow_forest_time", "rules_extraction_time"]]
    .std()
    .reset_index()
)
df_python
```

```python
#R
data = {
    "run": [1]*9 + [2]*9 + [3]*9 + [4]*9 + [5]*9,
    "n_dim": [15, 25, 50, 75, 100, 125, 150, 175, 200]*5,
    
    # Fit times
    "fit_time": time_r_dims.ravel().tolist(),
    
    # Grow forest times
    "grow_forest_time": GrowForest_dims_r.ravel().tolist(),
    
    # Rules extraction times
    "rules_extraction_time": RulesExtraction_dims_r.ravel().tolist(),
}

# Create dataframe
df_r = pd.DataFrame(data)

# Compute mean times across runs for each n_dim
df_r_summary_mean = (
    df_r
    .groupby("n_dim")[["fit_time", "grow_forest_time", "rules_extraction_time"]]
    .mean()
    .reset_index()
)
df_r_summary_std = (
    df_r
    .groupby("n_dim")[["fit_time", "grow_forest_time", "rules_extraction_time"]]
    .std()
    .reset_index()
)
df_r
```

```python
# Julia
n_dim = [15, 25, 50, 75, 100, 125, 150, 175, 200]

data = {
    "run": [1]*9 + [2]*9 + [3]*9 + [4]*9 + [5]*9,
    "n_dim": n_dim * 5,

    # R calls it fit_time -> map from Julia "training_time"
    "fit_time": time_julia_dims.ravel().tolist(),

    # Grow forest times (Run 1..5, each has 9 values)
    "grow_forest_time": GrowForest_dims_julia.ravel().tolist(),

    # Rules extraction times (Run 1..5, each has 9 values)
    "rules_extraction_time": RulesExtraction_dims_julia.ravel().tolist(),
}

df_jl = pd.DataFrame(data)

df_jl_summary_mean = (
    df_jl
    .groupby("n_dim")[["fit_time", "grow_forest_time", "rules_extraction_time"]]
    .mean()
    .reset_index()
)
df_jl_summary_std = (
    df_jl
    .groupby("n_dim")[["fit_time", "grow_forest_time", "rules_extraction_time"]]
    .std()
    .reset_index()
)

df_jl
```
